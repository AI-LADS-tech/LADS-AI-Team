```python
import matplotlib.pyplot as plt
import numpy as np 
xpoint=np.array([23,67])
ypoint=np.array([45,123])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
xpoint=np.array([1,5,7,1,9,1])
ypoint=np.array([1,9,1,7,7,1])
plt.plot(xpoint,ypoint , 'o:b', ms =20 , mec = 'r' , linestyle = 'dashed')
plt.show()
```

```python
xpoint=np.array([1,5,7,1,9,1])
ypoint=np.array([1,9,1,7,7,1])
plt.plot(xpoint,ypoint , marker = 'o', ms = 30, mec = 'c', mfc = 'r' , linewidth = '10')
plt.show()
```

```python
# multi line ploting
x1=np.array([1,2,5,3,6,7,4,8])
x2=np.array([9,4,3,7,7,4,8,9])
x3=np.array([4,9,5,3,6,8,4,8])
x4=np.array([1,5,5,4,6,7,9,8])
plt.plot(x1)
plt.plot(x2)
plt.plot(x3)
plt.plot(x4)
plt.show()
```

```python
#subploting concept
 #plot one 
x1=np.array([1,2,5,3,6,7,4,8])
x2=np.array([9,4,3,7,7,4,8,9])
plt.subplot(1,2,1)
plt.title("Plot 1")
plt.plot(x1,x2)

 #plot2
    
x3=np.array([4,9,5,3,6,8,4,8])
x4=np.array([1,5,5,4,6,7,9,8])
plt.subplot(1,2,2)
plt.title("Plot 2")
plt.plot(x3,x4)

plt.show()
```

```python
#plotting multiple plots
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.title("Plot 1")
plt.plot(x,y)
print("\n")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.title("Plot 2")
plt.plot(x,y)
print("\n")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.title("Plot 3")
print("\n")
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.title("Plot 4")
plt.plot(x,y)
print("\n")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.title("Plot 5")
plt.plot(x,y)
print("\n")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.title("Plot 6")
plt.plot(x,y)
print("\n")

plt.show()
```
